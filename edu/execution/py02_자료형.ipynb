```python
# 위치지정 물품추가
box = ['호박', '감자', '소주', '맥주', '피자', '치킨']
# 입력받기1: 추가할 물건 이름을 적어주세요
# 입력받기2: 어느 물품 뒤에 넣어둘까요
# box를 출력

itemName = input("추가할 물건 이름을 적어주세요")
where = input("어느 물품 뒤에 넣어둘까요")

box.insert(box.index(where)+1, itemName)

print(box)
```

### 딕셔너리 관련함수

```python
# Key 키 리스트 생성
a = {'name': 'pey', 'phone': '010-9999-1234', 'birth': "1118"}
ky = a.keys()
l_ky = list(ky)
print(a, type(a), ky, type(ky), type(l_ky), sep= "\n")
```

```python
# Values 키 리스트 생성
a = {'name': 'pey', 'phone': '010-9999-1234', 'birth': "1118"}
va = a.values()
l_va = list(ky)
print(a, type(a), va, type(va), type(l_va), sep= "\n")    
```

```python
# items 키값 쌍을 리스트 안의 튜플로 생성
a = {'name': 'pey', 'phone': '010-9999-1234', 'birth': "1118"}
it = a.items()
l_it = list(ky)
print(a, type(a), it, type(it), type(l_it), sep= "\n")    
```

```python
# clear - 키값 쌍 모두 지우기
a = {'name': 'pey', 'phone': '010-9999-1234', 'birth': "1118"}
a.clear()
print(a)
```

```python
# get - Key로 Value 얻기
a = {'name': 'pey', 'phone': '010-9999-1234', 'birth': "1118"}
print(a.get("name"))
print(a.get("ok"))
```

```python
# in - 해당 키 값 조사하기
a = {'name': 'pey', 'phone': '010-9999-1234', 'birth': "1118"}

print("name" in a)
print("phone" in a)
print("OK" in a)
```

```python
# pop - 가져와서 생략
a = {'name': 'pey', 'phone': '010-9999-1234', 'birth': "1118"}
pp = a.pop("phone")

print("꺼낸값", pp)
print("남은값", a)
```
